# Model training & testing

```python
from sklearn.model_selection import train_test_split

X1 = df_encoded.drop('Output', axis=1)
y1 = df_encoded['Output']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

clf1 = DecisionTreeClassifier(random_state=42)

skf = StratifiedKFold(n_splits=10)

scores = cross_val_score(clf1, X_train1, y_train1, cv=skf, scoring='accuracy')

clf1.fit(X_train1, y_train1)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf1, X_train1, y_train1, X_test1, y_test1, train=True):
    if train:
        pred = clf1.predict(X_train1)
        clf1_report = pd.DataFrame(classification_report(y_train1, pred, output_dict=True))
        print("Train Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(y_train1, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf1_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train1, pred)}\n")

    elif train==False:
        pred = clf1.predict(X_test1)
        clf1_report = pd.DataFrame(classification_report(y_test1, pred, output_dict=True))
        print("Test Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(y_test1, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf1_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test1, pred)}\n")

print_score(clf1, X_train1, y_train1, X_test1, y_test1, train=True)
print_score(clf1, X_train1, y_train1, X_test1, y_test1, train=False)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf2 = RandomForestClassifier(n_estimators=100)


skf = StratifiedKFold(n_splits=10)
scores = cross_val_score(rf_clf2, X_train1, y_train1, cv=skf, scoring='accuracy')

rf_clf2.fit(X_train1, y_train1)

scores
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report

bagged_tree = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, criterion='entropy'),
    random_state=42
)

skf = StratifiedKFold(n_splits=10)

scores = cross_val_score(bagged_tree, X_train1, y_train1, cv=skf, scoring='accuracy')
print(f"Cross-Validation Accuracy: {scores.mean():.2f} ± {scores.std():.2f}")

bagged_tree.fit(X_train1, y_train1)

y_pred = bagged_tree.predict(X_test1)


print("\nClassification Report:\n")
print(classification_report(y_test1, y_pred))

first_tree = bagged_tree.estimators_[0]
print(export_text(first_tree, feature_names=list(X_train1.columns)))
```
