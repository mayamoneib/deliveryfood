```python
encode = new_version.copy()
```

```python
categorical_columns = ['Marital Status', 'Occupation','Gender', 'Meal Type', 'Preferred Order Time', 'Check Reviews Before Delivery']
encode = pd.get_dummies(data=encode, columns=categorical_columns)
```

```python

from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
columns_to_order = [
    "Time-saving", "Delivery over Cooking", "Good App UI",
    "High-quality products", "Tasty Food", "Many Food Brands", "Easy Payment",
    "Accurate Tracking System", "Friendly delivery personnel", "Fresh food",
    "Appropriate temperature", "High-quality packaging", "Age", "Max Wait Time",
    "Area Type", 'Offers Promotions', 'Late Delivery', 'Poor Hygiene',
    'Unavailable Demanded Items', 'Unaffordable Prices', 'Insufficient delivery personnel',
    'Wrong delivery order', 'Missing items'
]

# Apply categorical ordering to each column
for col in columns_to_order:
   encode[col] = label.fit_transform(encode[col])
```

```python
from sklearn.model_selection import train_test_split

X4 = encode.drop('Purchase Decision',axis=1) # Independent variable
y4 = encode['Purchase Decision'] # Dependent variable

train_x, test_x, train_y, test_y = train_test_split(X4, y4, test_size=0.25, random_state=42)

print('Training dataset size:', train_x.shape[0])
print('Testing dataset size:', test_x.shape[0])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

clf4 = DecisionTreeClassifier(random_state=42)

# since we have a relatively small dataset (~1000 records),
# we'll use a low number of splits
skf = StratifiedKFold(n_splits=10)

# Note how cross validation is applied on the training dataset
# The testing dataset is always left out and used -only- in the final evaluation
scores = cross_val_score(clf4, train_x, train_y, cv=skf, scoring='accuracy')

clf4.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf4, train_x, train_y, test_x, test_y, train=True):
    if train:
        pred = clf4.predict(train_x)
        clf_report = pd.DataFrame(classification_report(train_y, pred, output_dict=True))
        print("Train Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(train_y, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(train_y, pred)}\n")

    elif train==False:
        pred = clf4.predict(test_x)
        clf_report = pd.DataFrame(classification_report(test_y, pred, output_dict=True))
        print("Test Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(test_y, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(test_y, pred)}\n")

# Call the function with the correct objects
print_score(clf4, train_x, train_y, test_x, test_y, train=True)
print_score(clf4, train_x, train_y, test_x, test_y, train=False)
```
