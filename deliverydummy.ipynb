```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#%matplotlib inline
#plt.style.use("seaborn")
#when using colab
# Set the style for plots using the updated seaborn style
plt.style.use("seaborn-v0_8") # Changed from 'seaborn' to 'seaborn-v0_8'

df = pd.read_csv('onlinedeliverydata.csv')
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.drop(['latitude', 'longitude', 'Pin code','Reviews', 'Monthly Income', 'Educational Qualifications',
              'Family size', 'Medium (P1)', 'Medium (P2)', 'Meal(P2)', 'Perference(P1)', 'Perference(P2)',
              'Health Concern', 'Bad past experience', 'Delay of delivery person picking up food', 'Google Maps Accuracy',
              'Good Road Condition', 'Low quantity low time', 'Delivery person ability', 'Number of calls'], axis=1, inplace=True)
```

```python
# Define the order of categories
ordered_categories = ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly agree']

# List of columns that need to be converted
columns_to_order = [
    "Ease and convenient", "Time saving", "More restaurant choices", "Easy Payment option",
    "More Offers and Discount", "Good Food quality", "Good Tracking system", "Self Cooking",
    "Late Delivery", "Poor Hygiene", "Unavailability",
    "Unaffordable", "Long delivery time", "Delay of delivery person getting assigned",
    "Wrong order delivered", "Missing item",
    "Order placed by mistake", "Residence in busy location"
]

# Apply categorical ordering to each column
for col in columns_to_order:
    df_copy[col] = pd.Categorical(df_copy[col], categories=ordered_categories, ordered=True)

# Verify changes
#print(df_copy.dtypes)  # Check the data types to confirm categorical conversion
df_copy[col]
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
columns_to_order = [
    "Ease and convenient", "Time saving", "More restaurant choices", "Easy Payment option",
    "More Offers and Discount", "Good Food quality", "Good Tracking system", "Self Cooking",
    "Late Delivery", "Poor Hygiene", "Unavailability",
    "Unaffordable", "Long delivery time", "Delay of delivery person getting assigned",
    "Wrong order delivered", "Missing item",
    "Order placed by mistake", "Residence in busy location"
]

# Apply categorical ordering to each column
for col in columns_to_order:
   df_copy[col] = label.fit_transform(df_copy[col])
```
