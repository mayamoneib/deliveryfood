```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.drop(['latitude', 'longitude', 'Pin code','Reviews'], axis=1, inplace=True)
```

```python
categorical_col = []
for column in df_copy.columns:
    if df_copy[column].dtype == object:
        categorical_col.append(column)
        print(f"{column} : {df_copy[column].unique()}")
        print("====================================")
```

specify the columns with 5-point likert scale as ordered categorical variables

```python
# Define the order of categories
ordered_categories = ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly agree']

# List of columns that need to be converted
columns_to_order = [
    "Ease and convenient", "Time saving", "More restaurant choices", "Easy Payment option",
    "More Offers and Discount", "Good Food quality", "Good Tracking system", "Self Cooking",
    "Health Concern", "Late Delivery", "Poor Hygiene", "Bad past experience", "Unavailability",
    "Unaffordable", "Long delivery time", "Delay of delivery person getting assigned",
    "Delay of delivery person picking up food", "Wrong order delivered", "Missing item",
    "Order placed by mistake", "Residence in busy location", "Google Maps Accuracy",
    "Good Road Condition", "Low quantity low time", "Delivery person ability"
]

# Apply categorical ordering to each column
for col in columns_to_order:
    df_copy[col] = pd.Categorical(df_copy[col], categories=ordered_categories, ordered=True)

# Verify changes
#print(df_copy.dtypes)  # Check the data types to confirm categorical conversion
df_copy[col]
```

```python
# Define the order of categories
ordered_categories = ['Unimportant', 'Slightly Important', 'Moderately Important', 'Important', 'Very Important']

# List of columns that need to be converted
columns_to_order = [
    "Less Delivery time", "High Quality of package", "Number of calls", "Politeness",
    "Freshness ", "Temperature", "Good Taste ", "Good Quantity"
]

# Apply categorical ordering to each column
for col in columns_to_order:
    df_copy[col] = pd.Categorical(df_copy[col], categories=ordered_categories, ordered=True)

# Verify changes
#print(df_copy.dtypes)  # Check the data types to confirm categorical conversion
df_copy[col]
```
